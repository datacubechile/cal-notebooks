# NASA DEM

The LP DAAC manages the distribution of NASA's NASADEM_HGT dataset, which provides global elevation data with 1 arc second spacing. It is derived from telemetry data collected during the SRTM mission, a collaboration between NASA, NGA, and German and Italian space agencies. NASADEM incorporates various data sources to improve accuracy, including ICESat GLAS ground control points and other satellite datasets. It covers approximately 80% of Earth's landmass between 60° N and 56° S latitude. The dataset includes layers such as DEM, NUM, and a water mask, and is available in NetCDF4 format as NASADEM_NC.
